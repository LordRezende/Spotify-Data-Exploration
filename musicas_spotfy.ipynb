```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.info()
```

**Realizando uma limpeza no dataset**

Retirando a coluna Unnamed: 0

```python
df = df.drop(columns=["Unnamed: 0"])
```

Verificando linhas que possuem colunas com valores nulos 

```python
df.isnull().sum()
```

Realizando um drop nas linhas encontradas

```python
df = df.dropna()
```

Verificação final

```python
df.isnull().sum()
```

**Filtrando valores duplicados**

Somando a quantidade de linhas duplicadas 

```python
print('Total de linhas duplicadas:', df.duplicated().sum())
```

Somando a quantidade de linhas com o ID duplicado

```python
print('Total de IDs duplicados:', df.duplicated(subset='track_id').sum())
```

Realizando um drop de todas as linhas duplicadas, deixando apenas aqueles com a maior popularidade  

```python
df = df.sort_values('popularity', ascending=False)
df = df.drop_duplicates(subset='track_id')
```

Verificação final

```python
print('Total de linhas duplicadas:', df.duplicated().sum())
print('Total de IDs duplicados:', df.duplicated(subset='track_id').sum())
```

```python
df['popularity'].describe()
```

**Primeira etapa: top 10 gêneros mais populares do Spotify**

Agrupando os gêneros por média de popularidade

```python
genre_popularity_mean = df.groupby('track_genre', as_index=False)['popularity'].mean().sort_values('popularity', ascending=False).head(10)
genre_popularity_mean
```

**Segunda etapa: top 10 artistas mais populares**

```python
artists_popularity_mean = df.groupby('artists', as_index=False)['popularity'].mean().sort_values('popularity', ascending=False).head(10)
artists_popularity_mean
```

Ao analisar a primeira amostra dos artistas mais populares, foi identificado um problema: algumas linhas possuem mais de um nome na coluna "artists", devido à presença de colaborações (feats).
Para tornar a análise mais precisa e justa, foi realizada uma transformação nos dados, separando os artistas em registros individuais.

```python
df['artists'] = df['artists'].str.split(';')
df = df.explode('artists')
artists_popularity_mean = df.groupby('artists', as_index=False)['popularity'].mean().sort_values('popularity', ascending=False).head(10)
artists_popularity_mean
```
